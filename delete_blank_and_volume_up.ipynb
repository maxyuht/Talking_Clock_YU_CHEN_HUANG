```python
import os
from pydub import AudioSegment
from pydub.silence import detect_nonsilent

# 文件夹路径
input_folder = '/Users/yuhantao/Desktop/TalkingClock/Talking_clock/English_audio'  # 输入文件夹路径
output_folder = '/Users/yuhantao/Desktop/TalkingClock/Talking_clock/English_audio的副本'  # 输出文件夹路径

# 确保输出文件夹存在
os.makedirs(output_folder, exist_ok=True)

# 定义静音的阈值 (dBFS) 和检测的最小静音长度 (ms)
silence_thresh = -80  # 小于这个dBFS的部分被认为是静音
min_silence_len = 200  # 静音的最小长度 (300ms)

# 要保留的静音部分长度
retain_silence_len = 50  # 100毫秒的静音部分

# 遍历文件夹中的所有音频文件
for filename in os.listdir(input_folder):
    if filename.endswith(".wav"):  # 可以根据需要扩展其他格式
        # 加载音频文件
        file_path = os.path.join(input_folder, filename)
        audio = AudioSegment.from_wav(file_path)

        # 检测音频中有声部分的位置
        non_silent_ranges = detect_nonsilent(audio, min_silence_len=min_silence_len, silence_thresh=silence_thresh)

        # 如果检测到有声部分，则提取有声段
        if non_silent_ranges:
            # 获取音频的第一个有声部分开始和最后一个有声部分结束的位置
            start_of_audio = non_silent_ranges[0][0]  # 第一个有声段开始
            end_of_audio = non_silent_ranges[-1][1]  # 最后一个有声段结束

            # 保留一段静音部分
            start_of_audio = max(0, start_of_audio - retain_silence_len)  # 开头多保留100ms静音
            end_of_audio = min(len(audio), end_of_audio + retain_silence_len)  # 结尾多保留100ms静音

            # 截取保留的部分
            processed_audio = audio[start_of_audio:end_of_audio]

            # 保存处理后的音频文件
            output_path = os.path.join(output_folder, filename)
            processed_audio.export(output_path, format="wav")
            print(f"Processed and saved: {output_path}")
```

```python
import os
from pydub import AudioSegment

# 定义输入和输出文件夹的列表
input_folders = ['/Users/yuhantao/Desktop/TalkingClock/Talking_clock/German_audio', '/Users/yuhantao/Desktop/TalkingClock/Talking_clock/English_audio', '/Users/yuhantao/Desktop/TalkingClock/Talking_clock/Chinese_audio', '/Users/yuhantao/Desktop/TalkingClock/Talking_clock/French_audio']  # 输入文件夹列表
output_folders = ['/Users/yuhantao/Desktop/TalkingClock/Talking_clock/German_audio的副本', '/Users/yuhantao/Desktop/TalkingClock/Talking_clock/English_audio的副本', '/Users/yuhantao/Desktop/TalkingClock/Talking_clock/Chinese_audio的副本','/Users/yuhantao/Desktop/TalkingClock/Talking_clock/French_audio的副本']  # 对应输出文件夹列表

# 目标音量的分贝数
target_dBFS = -20.0  # 将所有音频文件的音量调到 -20 dBFS

# 确保输出文件夹存在
for output_folder in output_folders:
    os.makedirs(output_folder, exist_ok=True)

# 遍历每个输入文件夹和对应的输出文件夹
for input_folder, output_folder in zip(input_folders, output_folders):
    for filename in os.listdir(input_folder):
        if filename.endswith(".wav"):  # 可以根据需要扩展其他格式
            # 加载音频文件
            file_path = os.path.join(input_folder, filename)
            audio = AudioSegment.from_wav(file_path)

            # 计算音量调整幅度
            change_in_dBFS = target_dBFS - audio.dBFS
            # 调整音量
            normalized_audio = audio.apply_gain(change_in_dBFS)

            # 保存处理后的音频文件
            output_path = os.path.join(output_folder, filename)
            normalized_audio.export(output_path, format="wav")
            print(f"Processed and saved: {output_path}")
```
